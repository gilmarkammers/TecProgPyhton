````python
# Criando as tabelas

import pandas as pd
import numpy as np

# Base de transações
transacoes = pd.DataFrame({
    'id_cliente': [1, 2, 1, 3, 2, 1, 3, 2],
    'tipo':       ['Crédito', 'Débito', 'Débito', 'Crédito', 'Crédito', 'Débito', 'Débito', 'Crédito'],
    'categoria':  ['Salário', 'Alimentação', 'Transporte', 'Salário', 'Lazer', 'Alimentação', 'Transporte', 'Salário'],
    'valor':      [3200.0, 150.0, None, 4500.0, 200.0, 80.0, None, 2800.0],
    'canal':      ['App', 'Agência', 'App', 'Internet Banking', 'App', 'Agência', 'App', 'Internet Banking']
})

# Base de cadastro de clientes
clientes = pd.DataFrame({
    'id_cliente': [1, 2, 3, 4],
    'nome':       ['Ana', 'Bruno', 'Carlos', 'Diana'],
    'estado':     ['SP', 'RJ', 'MG', 'SP'],
    'perfil':     ['Premium', 'Básico', 'Intermediário', 'Básico']
})
````

````python
# Visualizando as tabelas
print("Tabela de Transações:")
display(transacoes)
print("\nTabela de Clientes:")
display(clientes)
````

## Tarefas

### Tarefa 1 — Conheça os dados

Use `shape`, `dtypes`, `head()` e `isna().sum()` nas **duas** tabelas separadamente.

- Quantos clientes existem no cadastro? Resposta: 4 Clientes
- Quantas transações existem? Resposta: 8 Transações
- Alguma coluna possui valores nulos? Resposta: Coluna valor na tabela de Transações

---


````python
# Usando `shape`, `dtypes`, `head()` e `isna().sum()` nas **duas** tabelas separadamente.

print("Shape da tabela de clientes:")
print(clientes.shape)

print("\nTipos de dados da tabela de clientes:")
print(clientes.dtypes)

print("\nPrimeiras linhas da tabela de clientes:")
display(clientes.head())

print("\nValores nulos da tabela de clientes:")
print(clientes.isna().sum())

print("\nShape da tabela de transações:")
print(transacoes.shape)

print("\nTipos de dados da tabela de transações:")
print(transacoes.dtypes)

print("\nPrimeiras linhas da tabela de transações:")
display(transacoes.head())

print("\nValores nulos da tabela de transações:")
print(transacoes.isna().sum())  


````


### Tarefa 2 — Limpeza

Preencha os valores nulos da coluna `valor` com a **mediana geral** da coluna.  
Use `fillna()`.

Confirme que não existem mais nulos com `isna().sum()`.

---


````python
# Preenchendo valores nulos com a mediana geral da coluna

mediana_valor = transacoes['valor'].median()
transacoes['valor'] = transacoes['valor'].fillna(mediana_valor)
print("\nValores nulos após preenchimento na tabela de transações:")

# Confirmando que não há mais valores nulos
print(transacoes.isna().sum())
````


### Tarefa 3 — Cálculos com NumPy

Com os valores da coluna `valor` (já limpa), crie um array NumPy e calcule:

- Soma total das transações
- Valor médio
- Maior valor
- Menor valor


````python
# Criando um array Numpy a partir da coluna 'valor' e calculando soma total, a média, maior valor e menor valor.
valores_array = transacoes['valor']
soma_total = np.sum(valores_array)
media_valores = np.mean(valores_array)
maior_valor = np.max(valores_array)
menor_valor = np.min(valores_array)

print(f"\nSoma total dos valores: {soma_total}")
print(f"Média dos valores: {media_valores}")
print(f"Maior valor: {maior_valor}")
print(f"Menor valor: {menor_valor}")

````


---

### Tarefa 4 — Join entre as bases

Faça um `merge` entre `transacoes` e `clientes` usando `id_cliente` como chave.

- Faça um **inner join** e salve em `df_inner`
- Faça um **left join** partindo de `clientes` e salve em `df_left`
- Compare o `shape` dos dois resultados

Em uma célula Markdown, explique com suas palavras: **por que os shapes são diferentes?**

---


````python
# Merge das tabelas de transações e clientes
# Inner join 
df_inner = pd.merge(transacoes, clientes, on='id_cliente', how='inner')
print("\nTabela com inner join:")
display(df_inner)

# Left join com clientes à esquerda
df_left = pd.merge(clientes, transacoes, on='id_cliente', how='left')
print("\nTabela com left join:")
display(df_left)

# Comparando o shape dos dois resultados:
print("\nShape do DataFrame com inner join:", df_inner.shape)
print("Shape do DataFrame com left join:", df_left.shape)
````

Por que os shapes são diferentes?
Resposta: O inner join retorna apenas as linhas que têm correspondência em ambas as tabelas, enquanto o left join retorna todas as linhas da tabela da esquerda (clientes) e as correspondências da tabela da direita (transações),preenchendo com NaN onde não há correspondência.



### Tarefa 5 — Investigando o left join

No `df_left`, use `loc` para filtrar apenas as linhas onde `valor` está nulo.

- Qual cliente consta no cadastro mas nunca realizou uma transação?

---


````python
linhas_nulas = df_left.loc[df_left['valor'].isna()]
print("\nQual cliente consta no cadastro mas nunca realizou uma transação?:")
display(linhas_nulas)
````


### Tarefa 6 — Análise por perfil

Usando o `df_inner`, aplique `groupby` na coluna `perfil` e calcule a **soma** e a **média** dos valores por perfil de cliente.

- Qual perfil teve o maior valor médio por transação?


````python
# Usando o `df_inner`, aplique `groupby` na coluna `perfil` e calcule a **soma** e a **média** dos valores por perfil de cliente.

grouped = df_inner.groupby('perfil')['valor'].agg(['sum', 'mean'])
print("\nSoma e média dos valores por perfil de cliente:")
display(grouped)

# Qual perfil de cliente tem a maior média de valor transacionado?
# Resposta: O perfil Intermediario teve o maior valor medio por transação

````
