## 📦 1. Criação dos DataFrames

```python
import pandas as pd
import numpy as np

np.random.seed(42)

# Clientes
df_clientes = pd.DataFrame({
    'id_cliente': np.arange(1, 101),
    'nome': ['Cliente_' + str(i) for i in range(1, 101)],
    'idade': np.random.choice([25, 30, 35, 40, np.nan], 100),
    'cidade': np.random.choice(['RJ', 'SP', 'MG', None], 100),
    'status_conta': np.random.choice(['Ativa', 'Inativa', None], 100),
    'score_credito': np.random.choice([300, 500, 700, 900, np.nan], 100),
    'renda_mensal': np.random.choice([2000, 4000, 6000, 8000, np.nan], 100),
    'segmento': np.random.choice(['Varejo', 'Alta Renda', None], 100)
})

# Inserindo duplicados
df_clientes = pd.concat([df_clientes, df_clientes.iloc[:5]])

# Transações
df_transacoes = pd.DataFrame({
    'id_cliente': np.random.randint(1, 101, 200),
    'valor_transacao': np.random.choice([100, 500, 1000, 2000, np.nan], 200),
    'tipo': np.random.choice(['Crédito', 'Débito'], 200),
    'mes': np.random.choice(['Jan', 'Fev', 'Mar'], 200)
})

df_clientes.head()
```

## 🎯 MISSÃO DO ALUNO

### 🔎 PARTE 1 — Diagnóstico

**Pergunta 1**  
Quantos valores nulos existem por coluna?

**Pergunta 2**  
Existem registros duplicados?


```python
# Quantos valores nulos existem em cada coluna do DataFrame de clientes?
nulos_por_coluna = df_clientes.isnull().sum()
print(nulos_por_coluna)

# Quantos valores nulos existem em cada coluna do DataFrame de transações?
nulos_por_coluna_transacoes = df_transacoes.isnull().sum()
print(nulos_por_coluna_transacoes)

```

```python
# Pergunta 1: Existem registros duplicados no DataFrame de clientes? Se sim, quantos?

duplicados_clientes = df_clientes.duplicated().sum()
print(f"Registros duplicados no DataFrame de clientes: {duplicados_clientes}")

# Pergunta 2: Existem registros duplicados no DataFrame de transações? Se sim, quantos?

duplicados_transacoes = df_transacoes.duplicated().sum()
print(f"Registros duplicados no DataFrame de transações: {duplicados_transacoes}")
```

### 📊 PARTE 2 — Análise exploratória

**Pergunta 3**  
Quantas cidades diferentes existem?

**Pergunta 4**  
Distribuição de status da conta


```python
# Pergunta 3: Quantas cidades diferentes existem no DataFrame de clientes?
cidades_diferentes = df_clientes['cidade'].nunique()
print(f"Cidades diferentes no DataFrame de clientes: {cidades_diferentes}")
```

```python
# Pergunta 4: Distribuição de status de conta no DataFrame de clientes
distribuicao_status_conta = df_clientes['status_conta'].value_counts()
print("Distribuição de status de conta:")
print(distribuicao_status_conta)
```

### 🧹 PARTE 3 — Limpeza de dados

**Pergunta 5**  
Remova duplicados  

**Pergunta 6**  
Preencha valores faltantes:
- idade → mediana  
- renda → média  
- cidade → "Não Informado"  
- status → ffill  


```python
# Pergunta 5: Remova duplicados

df_clientes.drop_duplicates()
df_transacoes.drop_duplicates()
```

```python
# Pergunta 6: Preencha os valores faltantes na coluna 'idade' do DataFrame de clientes com a mediana das idades.
mediana_idade = df_clientes['idade'].median()
df_clientes['idade'] = df_clientes['idade'].fillna(mediana_idade)

# Preencha os valores faltantes na coluna renda_mensal do DataFrame de clientes com a média das rendas mensais.
media_renda_mensal = df_clientes['renda_mensal'].mean()
df_clientes['renda_mensal'] = df_clientes['renda_mensal'].fillna(media_renda_mensal)

# Preencha os valores faltantes na coluna cidade do DataFrame de clientes com o texto "Não informado".
df_clientes['cidade'] = df_clientes['cidade'].fillna("Não informado")

# Preencha os valores faltantes na coluna status_conta do DataFrame de clientes com o metodo forward fill.
df_clientes['status_conta'] = df_clientes['status_conta'].ffill()
```
