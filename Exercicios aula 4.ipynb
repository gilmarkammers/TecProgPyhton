```python
import pandas as pd
import numpy as np

print(pd.__version__)

#Criando uma série simples com indice padrao
precos = pd.Series([100, 200, 300, 400, 500])
print(precos)

#Criando uma série com indices personalizados
precos = pd.Series([100, 200, 300, 400, 500], index=['A', 'B', 'C', 'D', 'E'])
print(precos)

# Criando uma série a partir de um dicionário
dados = {'A': 100, 'B': 200, 'C': 300, 'D': 400, 'E': 500}
precos = pd.Series(dados)
print(precos)



```

### Exercício 1 — Criando e inspecionando uma Series

**Enunciado:** Crie uma Series com as temperaturas médias mensais de uma cidade no primeiro semestre do ano: Janeiro=28, Fevereiro=29, Março=27, Abril=25, Maio=23, Junho=21. Use os **nomes dos meses** como índice. Em seguida:
- Exiba a temperatura de Março.
- Exiba as temperaturas de Fevereiro a Abril.
- Calcule a média do semestre com `.mean()`.

> 💡 **Dica:** Crie a Series com `pd.Series({...})` usando um dicionário, ou com `pd.Series([...], index=[...])` passando a lista de índices separada.

```python
temperaturas = pd.Series([28, 29, 27, 25, 23, 21], index=['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho  '])
print(temperaturas)

print("Temperatura de Março:", temperaturas['Março'])
print("Temperaturas de Fevereiro a Abril:", temperaturas['Fevereiro':'Abril'])
print("Média do semestre:", temperaturas.mean())
```

### Exercício 2 — Criando e explorando um DataFrame

**Enunciado:** Crie um DataFrame com os dados de 5 filmes brasileiros:

| titulo | ano | genero | nota_imdb |
|--------|-----|--------|-----------|
| Cidade de Deus | 2002 | Drama | 8.6 |
| Tropa de Elite | 2007 | Ação | 8.0 |
| Central do Brasil | 1998 | Drama | 7.8 |
| O Auto da Compadecida | 2000 | Comédia | 8.2 |
| Bacurau | 2019 | Suspense | 7.4 |

Exiba: as 3 primeiras linhas, o `shape`, os `dtypes` e as estatísticas com `describe()`.

> 💡 **Dica:** Use um dicionário de listas para criar o DataFrame. Lembre que cada chave é o nome da coluna e cada lista são os valores.

```python
filmes = pd.DataFrame({
    'Título': ['Cidade de Deus', 'Tropa de Elite', 'Central do Brasil', 'O Auto da Compadecida', 'Bacurau'],
    'Ano': [2002, 2007, 1998, 2000, 2019],
    'Gênero': ['Drama', 'Ação', 'Drama', 'Comédia', 'Suspense'],
    'nota_imdb': [8.6, 8.0, 7.8, 8.2, 7.4]
})

print(filmes[0:3])

print(filmes.shape)

print(filmes.dtypes)

print(filmes.describe())

```
